```python
from score import ScoreFile
from pandas import read_csv

data = read_csv("score.sc",delim_whitespace=True,header=1) # Note, the header may not be the first line
sf = ScoreFile(data)
print sf
```

```python
print sf[ sf.total_score < -200 ]
#gives rows with total_score below a certain cutoff -- not necessarily a certain percentage of total number of decoys
```

```python
sfSort = sf.sort("total_score")
print sfSort
#sorted based on total_score, with lowest total_score at the top
```

```python
ten_perc = int(.10*sfSort.shape[0])
top_ten = sfSort.head(n=ten_perc)
#print top_ten
print top_ten.description
#output is the top ten percent of decoys with the lowest protein energy
```

```python
list_top_ten = []
temp_top = top_ten
for index in range (top_ten.shape[0]):
    decoy = str(temp_top.head(n=1).description).split("\n")[0].split(" ")[-1]
    temp_top = temp_top.drop(temp_top.index[:1])
    list_top_ten.append(decoy)
    
print list_top_ten
#list of the top ten decoys -- there could probably be a better, more efficient way to do this.
```

```python
#takes the decoy_file, and extracts out just the line with the ligand energies; returns that line as a string
def extract_ligand_energy(decoy_file):
    string_match = "END_POSE_ENERGIES_TABLE"
    decoy_file_list = decoy_file.readlines()
    for line_index in range (len(decoy_file_list)):
        if string_match in decoy_file_list[line_index]:
            ligand_energy_line = decoy_file_list[line_index-1]
            print ligand_energy_line
    return(ligand_energy_line)
```

```python
def extract_total_ligand_energy(ligand_energy_line):
    return ligand_energy_line.strip().split(" ")[-1]
```

```python
#want to cycle through the list of top decoys, append .pdb extension, read the file, and extract the ligand energy from
#the end of the file...
extension = ".pdb"
ligand_energy_list = []
for decoy in list_top_ten:
    file_name = decoy+extension
    decoy_file = open(file_name, "r")
    ligand_energy = extract_ligand_energy(decoy_file)
    ligand_energy_list.append((file_name, float(extract_total_ligand_energy(ligand_energy))))
    
    print ligand_energy_list
    
```

```python
#sort the list of ligand energies, and then take the lowest energy as the best decoy
sorted_ligand_energies = sorted(ligand_energy_list, key=lambda x: x[1])
print sorted_ligand_energies
best_decoy = sorted_ligand_energies[0]
print "best decoy = ", best_decoy
```
